```python
a=np.zeros(len(flist119[:200]))
print (len(a))
for i,f in enumerate(flist119[:200]):
    a[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1000:13500,1000:1350])
print (a.mean())
print (a.std())


flux0=(a-a.mean())/a.std()
flux=flux0.copy()
```

```python
pl.figure(figsize=(15,5))
pl.plot(flux)
pl.xlabel("img number")
pl.ylabel("standardized flux")
```

```python
b=np.zeros(len(flist120[:200]))
print (len(a))
for i,f in enumerate(flist120[:200]):
    b[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1000:13500,1000:1350])
print (b.mean())
print (b.std())


flux2=(b-b.mean())/b.std()
pl.figure(figsize=(15,5))
pl.plot(flux2)
pl.xlabel("img number")
pl.ylabel("standardized flux")
```

```python
c=b.copy()
res=np.polyfit(np.arange(len(a)),c,1)
print (res)
c-=(np.arange(len(a))*res[0]+res[1])
flux3=(c-c.mean())/c.std()
pl.figure(figsize=(15,5))
pl.plot(flux3)
pl.xlabel("img number")
pl.ylabel("standardized flux")
```

```python
d=np.zeros(len(flist81[:200]))
print (len(a))
for i,f in enumerate(flist81[:200]):
    d[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1000:13500,1000:1350])
print (d.mean())
print (d.std())


flux5=(d-d.mean())/d.std()
pl.figure(figsize=(15,5))
pl.plot(flux5)
pl.xlabel("img number")
pl.ylabel("standardized flux")
```

```python
from gatspy import  periodic
model119 = periodic.LombScargleFast(fit_period=True)
model119.optimizer.period_range = (0.2, 3.0)
model119.fit(np.arange(len(a))/4.0*10.0,flux)
periods119, power119 = model119.periodogram_auto()
print ("best period", model119.best_period)
```
