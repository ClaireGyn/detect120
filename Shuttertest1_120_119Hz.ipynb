```python
a=np.zeros(len(flist119[:200]))
print (len(a))
for i,f in enumerate(flist119[:200]):
    a[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1000:13500,1000:1350])
print (a.mean())
print (a.std())
flux=(a-a.mean())/a.std()
pl.figure(figsize=(15,5))
pl.plot(flux)
pl.xlabel("img number")
pl.ylabel("almost normalized flux")

b=np.zeros(len(flist120[:200]))
print (len(a))
for i,f in enumerate(flist120[:200]):
    b[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1000:13500,1000:1350])
print (b.mean())
print (b.std())
flux2=(b-b.mean())/b.std()
pl.figure(figsize=(15,5))
pl.plot(flux2)
pl.xlabel("img number")
pl.ylabel("almost normalized flux")
```

```python
model119 = LombScargleFast().fit(np.arange(len(a))/4.0*10.0,flux)
model120 = LombScargleFast().fit(np.arange(len(a))/4.0*10.0,flux2)
periods119, power119 = model119.periodogram_auto()
periods120, power120 = model120.periodogram_auto()
```
