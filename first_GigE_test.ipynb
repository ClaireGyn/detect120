aperture photometry gives us the flux for the ~1000 images. 

```python
a=np.zeros(len(flistgood))
for i,f in enumerate(flistgood):
    a[i]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol,nband)[1100:1500,1400:1600])
```

 normalized and standardized so that it its mean 0 and span 1

```python
flux=a/a.mean()
print a.mean()
print a.std()
flux-=1.0
flux/=0.008 
pl.figure(figsize=(15,5))
pl.plot(flux)
pl.xlabel("img number")
pl.ylabel("almost normalized flux")
```

we see a wave, yay!! but it is puzzling: very irregular. may be due to shakes, as I have not registered the images. 

what is our sampling: what is that in physical time (seconds)? if i am right about our cadence this looks like this

```python
runtime=np.arange(len(a))/8.0*10.0 #8Hz, but pick one every 10
pl.figure(figsize=(15,5))
pl.plot(runtime,flux)
pl.xlabel("seconds")
pl.ylabel("almost normalized flux")
```

the epected frequency is 120Hz

```python
freq=120
phase=0
wave = lambda t,phi,freq,amp: np.sin(2.*freq*np.pi*t+phi)*amp
t=np.linspace(0,200,350000)
```

```python
pl.figure(figsize=(15,5))
pl.plot(t,wave(t,phase,freq,1), '-',color='SteelBlue', alpha=0.5)
pl.plot(runtime,flux,'o',markeredgecolor='IndianRed',markerfacecolor='None')
pl.xlabel("seconds")
pl.ylabel("almost normalized flux")
pl.xlim(0,0.1)
```

ok, this plot that overlays the expected frequency with the observe flux tells me nothing. let's sample the 120Hz frequency at what i expect my sampling to be: 0.8Hz (8 Hz, pick 1 out of 10)

```python
tsample=np.linspace(200,600,400*8/10)
pl.figure(figsize=(25,5))
pl.plot(tsample,wave(tsample,phase,freq,1), '-',color='SteelBlue', alpha=0.5)
pl.plot(tsample,wave(tsample,phase,freq,1), 'o',color='SteelBlue', alpha=0.5)
pl.plot(runtime,flux,'o',markeredgecolor='Red',markerfacecolor='None', lw=2)
pl.xlabel("seconds")
pl.ylabel("almost normalized flux")
pl.xlim(0,600)
tsample=np.linspace(600,1000,400*8/10)
pl.figure(figsize=(25,5))
pl.plot(tsample,wave(tsample,phase,freq,1), '-',color='SteelBlue', alpha=0.5)
pl.plot(tsample,wave(tsample,phase,freq,1), 'o',color='SteelBlue', alpha=0.5)
pl.plot(runtime,flux,'o',markeredgecolor='Red',markerfacecolor='None', lw=2)
pl.xlabel("seconds")
pl.ylabel("almost normalized flux")
pl.xlim(600,1200)
```

**that is NOT the right frequency we are sampling....**

```python
from scipy import optimize
def gauss(mu,sigma):
    return 1.0/(2.0*np.pi*sigma)*np.exp(-(runtime-mu)**2*0.5**2/sigma**2)
    

def fitsampling((freq,phase),x,y,amp):
    diffs=wave(x,phase,freq*amp)-y
    return sum(diffs)**2

pl.plot(runtime, gauss(runtime.mean(),10.))
pl.figure(figsize=(15,15))
dflux=numpy.convolve(np.diff(flux),gauss(runtime.mean(),5.),mode='same')
pl.plot(runtime,flux)
pl.plot(runtime,dflux)
pl.plot(runtime[1:][(dflux[1:]<0)*(dflux[:-1]>0)],dflux[1:][(dflux[1:]<0)*(dflux[:-1]>0)],'o')

i0=0
pars0=[1./60,3*np.pi*0.5]
for i in np.where((dflux[1:]<0) & (dflux[:-1]>0))[0]:
    pl.figure()
    pars0[0]=0.5/(runtime[i]-runtime[i0])
    pars=optimize.minimize(fitsampling, pars0, args=(runtime[i0:i],flux[i0:i]), method='Powell')#, options={'maxiter':4})
    pl.plot(runtime[i0:i],flux[i0:i],'bo')
    pl.plot(runtime[i0:i],wave(runtime[i0:i],pars.x[1],pars.x[0],amp=1),'k-')
    pl.plot(runtime[i0:i],wave(runtime[i0:i],0,pars0[0]),'r-')
    i0=i
    pars0[1]=0
    print pars
```
