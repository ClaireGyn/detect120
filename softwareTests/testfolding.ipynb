```python
import pylab as pl
import numpy as np
%pylab inline

wave = lambda freq,t: np.sin(2.*freq*np.pi*t)

def folding (fluxes, timestamps, frequency, ax):
    print (timestamps)
    folded = {}
    #print (timestamps)
    #print (timestamps%(2/float(frequency)))
    runtime= (timestamps)%(2./float(frequency))
    print (runtime )
    sorted_idx = np.array([i[0] for i in sorted(enumerate(runtime), key=lambda x:x[1])])
    folded['time'] = runtime[sorted_idx]
    folded['flux'] = fluxes[sorted_idx]
    
    ax.plot(timestamps, fluxes)
    ax.plot(folded['time'], folded['flux'], '-', color='deeppink', linewidth=0.75)
    ax.plot(folded['time'], folded['flux'], 'o', color='deeppink')
    ax.set_xlabel("Microseconds", fontsize=15)
    ax.set_ylabel("Almost Normalized Flux (Folded)", fontsize=15)
    ax.set_title('test, true freq = %d, sampling = %f'%(frequency, timestamps[1]-timestamps[0]) , fontsize=20)
    #pl.xlim(0, 2*1/float(frequency)*1000)
```
