this is the first attempt to detect the 120Hz phase of the electrical grid. 
We use the PointGray GigE camera, 5Mpix color, relatively high cadence enabled.
In this forst attempt we point it to on e of the CUSP ceiling lights, the set up is shady: we do not have a proper tripot so we ductape the camera to some pedistal.
we ran the camera at:

5ms integration, 40ms integration 100ms integrations, and at 6,7,1.1,1 Hz

longer integration would couse saturation. shorter integration or higher frequency would start getting us in trouble with the data buffering

```python
import glob
import numpy as np
import pylab as pl
from mytools import sigma_clip
%pylab inline

wave = lambda t,phi,freq,amp: np.sin(2.*freq*np.pi*t+phi)*amp
```

the images are 2048x2448, and i turned the color off, so i do not need a third axis, like i did before.

```python
nrow = 2048
ncol = 2448
flist={}
for freq in ['6', '7' ,'1.1', '1']:
    for exp in [5,40,100]:
        flist['%d_%s'%(exp,freq)]=glob.glob("/home/fb55/UIdata/test*%dms%shz*.raw"%(exp,freq.replace('.','_')))
        flist['%d_%s'%(exp,freq)]=np.array(flist['%d_%s'%(exp,freq)])[np.argsort(flist['%d_%s'%(exp,freq)])]
        print 'Exposure',exp,'ms',freq,'Hz','N images:',len(flist['%d_%s'%(exp,freq)])
#print flistgood
```

we collected ~1000 images in every mode, 

```python
fig=pl.figure(figsize=(15,15))
i=1
for exp in [5,40,100]:
    for freq in ['6', '7' ,'1.1', '1']:
        if len(flist['%d_%s'%(exp,freq)])>0:
            ax=fig.add_subplot(4,2,i)
            pl.imshow(np.fromfile(flist['%d_%s'%(exp,freq)][0],dtype=np.uint8).reshape(nrow,ncol))
            pl.title ('%dms %shz'%(exp,freq))
            i=i+1
```

we can zoom in and onto the light we have a fairly large area of 120x150 pixels that is far from the edges and gives us decent SNR
aperture photometry gives us the flux for the ~1000 images. 

```python
y1,yn,x1,xn=1020,1140,800,950
fig=pl.figure(figsize=(15,15))
i=1
a={}

for exp in [5,40,100]:
  for freq in ['6', '7' ,'1.1', '1']:
      if len(flist['%d_%s'%(exp,freq)])>0:
            print '%dms %shz'%(exp,freq)
            ax=fig.add_subplot(4,2,i)
            pl.imshow(np.fromfile(flist['%d_%s'%(exp,freq)][0],dtype=np.uint8).reshape(nrow,ncol)[y1:yn,x1:xn])
            pl.title ('%dms %shz'%(exp,freq))
            i=i+1
            a['%d_%s'%(exp,freq)]=np.zeros(1000)
            for j,f in enumerate(flist['%d_%s'%(exp,freq)][:1000]):
                a['%d_%s'%(exp,freq)][j]=sum(np.fromfile(f,dtype=np.uint8).reshape(nrow,ncol)[y1:yn,x1:xn])
```

 normalized and standardized so that it its mean 0 and span 1, and 3sigma clipped to remove greg walking in front of the camera (Really??)

```python
i=1
flux={}
fig=pl.figure(figsize=(15,5*len(a.keys())))
for exp in [5,40,100]:
    for freq in ['6', '7' ,'1', '1_1']:

        try:
            flux['%d_%s'%(exp,freq)]=sigma_clip(a['%d_%s'%(exp,freq)], sig=3, method='median', iterations=100, verbose=False)
            print "mean=",flux['%d_%s'%(exp,freq)].mean(),
            print "std=",flux['%d_%s'%(exp,freq)].std()
            ax=fig.add_subplot(len(a.keys()),1,i)
            flux['%d_%s'%(exp,freq)]=flux['%d_%s'%(exp,freq)]/flux['%d_%s'%(exp,freq)].mean()
            flux['%d_%s'%(exp,freq)]-=1.0 
            flux['%d_%s'%(exp,freq)]/=flux['%d_%s'%(exp,freq)].std()*1.5

            pl.plot(flux['%d_%s'%(exp,freq)], '-',color='SteelBlue')
            pl.title ('%dms %shz'%(exp,freq))
            pl.xlabel("img number")
            pl.ylabel("almost normalized flux")
            i+=1
        except KeyError: pass
```

clearly a wave is detected in ALL images, including the 5ms exposusre! there are some long term trends 40ms 6Hz. they may be due to external light changes, though i do not recall the sky being very variable, but also to the camera temperature changes perhaps.
